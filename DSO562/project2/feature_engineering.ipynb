## this is a sample trial

### for the convenience of checking the result, we use `date` to demonstrate the performance

```python
dic = {}
for lag in [3]:
    start = time.time()  
    for i in range(len(dates)):      
        if i == 0:   
            for f in ['date']:    
                present = subset.query('date_index== @i')[f]
                temp_counter = Counter()
                temp_count = np.array([])
                for p in present:
                    temp_counter.update([p])
                    temp_count = np.append(temp_count,temp_counter[p])
                dic[f+'_'+str(lag)] = np.array([]) #here we intialize the dictionary
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],temp_count))
        elif 1 <= i <= lag-2:
            temp = subset.query('date_index <= @i-1')
            for f in ['date']:        
                present = subset.query('date_index== @i')[f]
                temp_counter = Counter(temp[f])
                temp_count = np.array([])
                for p in present:
                    temp_counter.update([p])
                    temp_count = np.append(temp_count,temp_counter[p])
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],temp_count))
        else:
            temp = subset.query('@i-@lag+1 <= date_index <= @i-1') 
            for f in ['date']:        
                present = subset.query('date_index== @i')[f]
                temp_counter = Counter(temp[f])
                temp_count = np.array([])
                for p in present:
                    temp_counter.update([p])
                    temp_count = np.append(temp_count,temp_counter[p])
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],temp_count))

    end = time.time()
    print('lag{}, time elapsed: {}'.format(str(lag),str(end-start)))
```

## Now we know the trial works, we can scale it up

```python
features = ['ssn','zip5','dob','address','homephone','firstname_lastname',
           'homephone_address','homephone_zip5','address_zip5',
           'ssn_address','ssn_dob','ssn_zip5','ssn_homephone',
           'firstname_lastname_dob','firstname_lastname_ssn',
           'firstname_lastname_homephone','firstname_lastname_address',
           'firstname_lastname_zip5','address_zip5_homephone',
            'firstname_lastname_ssn_dob']
```

```python
lag_list = [1,3,7,15,30]
```

```python
dic = {}
for lag in lag_list:
    start = time.time()  
    for i in range(len(dates)):      
        if i == 0:   
            for f in features:    
                present = subset.query('date_index== @i')[f]
                temp_counter = Counter()
                temp_count = np.array([])
                for p in present:
                    temp_counter.update([p])
                    temp_count = np.append(temp_count,temp_counter[p])
                dic[f+'_'+str(lag)] = np.array([]) #here we intialize the dictionary
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],temp_count))
        elif 1 <= i <= lag-2:
            temp = subset.query('date_index <= @i-1')
            for f in features:        
                present = subset.query('date_index== @i')[f]
                temp_counter = Counter(temp[f])
                temp_count = np.array([])
                for p in present:
                    temp_counter.update([p])
                    temp_count = np.append(temp_count,temp_counter[p])
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],temp_count))
        else:
            temp = subset.query('@i-@lag+1 <= date_index <= @i-1') 
            for f in features:        
                present = subset.query('date_index== @i')[f]
                temp_counter = Counter(temp[f])
                temp_count = np.array([])
                for p in present:
                    temp_counter.update([p])
                    temp_count = np.append(temp_count,temp_counter[p])
                dic[f+'_'+str(lag)] = np.hstack((dic[f+'_'+str(lag)],temp_count))

    end = time.time()
    print('lag{}, time elapsed: {}'.format(str(lag),str(end-start)))
```
